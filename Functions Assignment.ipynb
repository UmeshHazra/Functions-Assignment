**1. What is the difference between a function and a method in Python?**

ANS:- A **function** is a block of code that is defined outside of any class and can be called independently.

A **method** is a function that is associated with an object (inside a class).

Example:-



```python
# Function
def greet():
    print("Hello")

greet()

# Method
class Person:
    def say_hello(self):
        print("Hello")

p = Person()
p.say_hello()
```

**2. Explain the concept of function arguments and parameters in Python.**

ANS:- Parameters are variables listed inside the parentheses in the function definition.

Arguments are the actual values passed to the function when calling it.

Example:-


```python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

**3. What are the different ways to define and call a function in Python?**

ANS:- Define a function using def keyword.

Call a function by writing its name followed by parentheses.

Example:-


```python
def greet(name):
    print(f"Hello {name}")

greet("Alice")

square = lambda x: x * x
print(square(4))
```

**4. What is the purpose of the return statement in a Python function?**

ANS:- The return statement is used to send back the result from a function to the caller.

Exaple:-

```python
def multiply(x, y):
    return x * y

product = multiply(4, 5)
print(product)
```

**5. What are iterators in Python and how do they differ from iterables?**

ANS:- **Iterable**: Any object you can loop over (like a list, tuple, string).

**Iterator**: An object which implements __iter__() and __next__() methods.

Example:-

```python
nums = [1, 2, 3]
it = iter(nums)

print(next(it))
print(next(it))
```

**6. Explain the concept of generators in Python and how they are defined.**
ANS:- **Generators** are special functions that yield values one at a time instead of returning all at once. They use the yield keyword.

Example:-

```python
def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
print(next(gen))
print(next(gen))
```

**7. What are the advantages of using generators over regular functions?**

ANS:-
Generators save memory because they yield one value at a time.

Useful for large datasets where holding everything in memory is expensive.

Example:-

```python
def numbers():
    for i in range(1000):
        yield i

gen = numbers()
print(next(gen))
```

**8. What is a lambda function in Python and when is it typically used?**

ANS:-
A lambda function is a small anonymous function defined using the lambda keyword.

Used for short, simple functions where defining a full def function is unnecessary.

Example:-

```python
add = lambda x, y: x + y
print(add(3, 4))
```

**9. Explain the purpose and usage of the map() function in Python.**

ANS:-
The map() function applies a given function to every item of an iterable (like a list).

Example:-

```python
nums = [1, 2, 3, 4]
squared = list(map(lambda x: x * x, nums))
print(squared)
```

**10. What is the difference between map(), reduce(), and filter() functions in Python?**

ANS:-
**map():** Applies a function to all items and returns the results.

**filter():** Applies a function that returns True/False and keeps only the items for which the function is True.

**reduce():** Repeatedly applies a function to the elements and reduces them to a single value.

```python
from functools import reduce

nums = [1, 2, 3, 4]

# map
mapped = list(map(lambda x: x * 2, nums))
print(mapped)

# filter
filtered = list(filter(lambda x: x % 2 == 0, nums))
print(filtered)

# reduce
summed = reduce(lambda x, y: x + y, nums)
print(summed)
```
